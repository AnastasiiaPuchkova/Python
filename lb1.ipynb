#### Флаг

```python
from IPython.display import HTML
import math

black = "#000000"
white = "#FFFFFF"
green = "#007A3D"
red = "#CE1126"

width = 600
height = 300
stripe_height = height / 3

x1, y1 = 0, 0
x2, y2 = 0, height
x3, y3 = width / 2, height / 2

a = math.sqrt((x3 - x2)**2 + (y3 - y2)**2) 
b = math.sqrt((x3 - x1)**2 + (y3 - y1)**2) 
c = math.sqrt((x2 - x1)**2 + (y2 - y1)**2) 

# Координати перетину бісектрис
star_center_x = (a * x1 + b * x2 + c * x3) / (a + b + c)
star_center_y = (a * y1 + b * y2 + c * y3) / (a + b + c)

# Розмір зірки: Діаметр круга = 1/14 довжини прапору
star_diameter = width / 14
star_radius_outer = star_diameter / 2 # Зовнішній радіус зірки
# Збільшення внутрішнього радіусу
star_radius_inner = star_radius_outer / 2.0

points = []
num_points = 7
start_angle = -90

for i in range(num_points):
    # Зовнішня точка
    outer_angle = math.radians(i * (360 / num_points) + start_angle)
    points.append(f"{star_center_x + star_radius_outer * math.cos(outer_angle):.2f},{star_center_y + star_radius_outer * math.sin(outer_angle):.2f}")

    # Внутрішня точка
    inner_angle = math.radians(i * (360 / num_points) + start_angle + (360 / (num_points * 2)))
    points.append(f"{star_center_x + star_radius_inner * math.cos(inner_angle):.2f},{star_center_y + star_radius_inner * math.sin(inner_angle):.2f}")

star_points = " ".join(points)

svg_code = f"""
<svg width="{width}" height="{height}" xmlns="http://www.w3.org/2000/svg">
  <rect x="0" y="0" width="{width}" height="{stripe_height}" fill="{black}" />
  <rect x="0" y="{stripe_height}" width="{width}" height="{stripe_height}" fill="{white}" />
  <rect x="0" y="{stripe_height * 2}" width="{width}" height="{stripe_height}" fill="{green}" />
  <polygon points="0,0 0,{height} {width/2},{height/2}" fill="{red}" />
  <polygon points="{star_points}" fill="{white}" />
</svg>
"""

display(HTML(svg_code))
```

### Функція:

```python
def find_primes_in_range(a, b):

    if not isinstance(a, int) or not isinstance(b, int):
        print("Помилка: введіть цілі числа.")
        return []
    
    if a > b:
        print("Помилка: перше число (a) має бути меншим або рівним другому (b).")
        return []

    prime_numbers = []
    
    for num in range(a, b + 1):
        is_prime_flag = True
        
        if num <= 1:
            is_prime_flag = False
        else:
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    is_prime_flag = False
                    break
        
        if is_prime_flag:
            prime_numbers.append(num)
            
    return prime_numbers
```

### Тести:

```python
find_primes_in_range(3, 3)
```
